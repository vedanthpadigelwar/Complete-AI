```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv('train.csv')
```

```python
train.shape
```

```python
train.head()
```

```python
iid = list(train['uniqueid'])
country = list(train['country'])
keys = []
for j,i in enumerate(iid):
    keys.append(str(i)+' x ' + str(country[j]))
```

```python
train.uniqueid = keys
train.head()
```

```python
cols = ['uniqueid','country', 'year', 'location_type',
       'cellphone_access', 'household_size', 'age_of_respondent',
       'gender_of_respondent', 'relationship_with_head', 'marital_status',
       'education_level', 'job_type', 'bank_account']
```

```python
train = train[cols]
train.head()
```

```python
enc = LabelEncoder()
train.bank_account = enc.fit_transform(train.bank_account)
train.head()
```

```python
train.uniqueid = (train.uniqueid).astype(str)
train.head()
```

```python
train.isna().any()
```

```python
train.isnull().any()
```

```python
train.bank_account = (train.bank_account).astype(int)
train.head()
```

```python
train.bank_account = (train.bank_account).astype(str)
train.head()
```

```python
seed = 42
X_train, X_test, y_train, y_test = train_test_split(train, train.bank_account, test_size=0.40, random_state=seed)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test )
```

```python
train.bank_account.value_counts()
```

```python
train = X_train
train.head()
```

```python
test = X_test.drop(columns=['bank_account'])
test.head()
```

```python
answer_key = pd.DataFrame(data=test.uniqueid,columns=['uniqueid'])
answer_key['bank_account'] = y_test
answer_key.head()
```

```python
type((answer_key.uniqueid).iloc[2])
```

```python
answer_key.bank_account.value_counts()
```

```python
sample_submission = answer_key.copy()
sample_submission.bank_account = str(1)
sample_submission.head()
```

```python
type((sample_submission.bank_account).iloc[2])
```

```python
sample_submission.bank_account.value_counts()
```

```python
train.to_csv('train.csv',index=False)
test.to_csv('test.csv',index=False)
sample_submission.to_csv('sample_submission.csv',index=False)
answer_key.to_csv('answer_key.csv',index=False)
```
